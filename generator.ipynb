```python
# -*- coding: utf-8 -*-
import pythoncom
from win32com.client import Dispatch, gencache

# Подключение API Компас 3D
kompas6_constants = gencache.EnsureModule("{75C9F5D0-B5B8-4526-8681-9903C567D2ED}", 0, 1, 0).constants
kompas6_constants_3d = gencache.EnsureModule("{2CAF168C-7961-4B90-9DA2-701419BEEFE3}", 0, 1, 0).constants
kompas_api7 = gencache.EnsureModule("{69AC2981-37C0-4379-84FD-5DD2F3C0A520}", 0, 1, 0)

# Получаем приложение Компас
application = kompas_api7.IApplication(Dispatch("Kompas.Application.7"))
Documents = application.Documents

# Основной цикл — создаем 3 модели
for i in range(3):
    # Параметры окружности
    center_x = i * 30        # X-координата центра
    center_y = 0             # Смещение по Y (вдоль оси Y — это высота от плоскости XOZ)
    center_z = 0             # Z-координата центра
    radius = 10 + i * 5      # Радиус окружности

    # Создаем новый документ
    doc = Documents.AddWithDefaultSettings(kompas6_constants.ksDocumentPart, True)
    doc3D = kompas_api7.IKompasDocument3D(doc)
    part = doc3D.TopPart

    # Получаем базовую плоскость и создаем смещённую
    part.GetEnt
    basePlane = part.GetDefaultEntity(kompas6_constants_3d.o3d_planeXOZ)
    offsetPlane = part.AddOffset(basePlane, center_y, True)

    # Создаем эскиз на смещённой плоскости
    sketchEntity = part.NewEntity(kompas6_constants_3d.o3d_sketch)
    sketchDef = sketchEntity.GetDefinition()
    sketchDef.SetPlane(offsetPlane)
    sketchEntity.Create()

    # Рисуем окружность в эскизе
    sketch2D = sketchDef.BeginEdit()
    sketch2D.ksCircle(center_x, center_z, radius, 1)
    sketchDef.EndEdit()

    # Создаем тело вращением
    rotateEntity = part.NewEntity(kompas6_constants_3d.o3d_bossRotated)
    rotateDef = rotateEntity.GetDefinition()
    rotateDef.SetSketch(sketchEntity)

    rotateParams = rotateDef.RotatedParam()
    rotateParams.direction = kompas6_constants_3d.dtNormal
    rotateParams.angleNormal = 360
    rotateParams.angleReverse = 0
    rotateParams.toroidShape = False  # True — если нужен тор

    rotateEntity.Create()

    # Сохраняем модель
    path = fr"C:\u4ebaENG\Diplom-\DSet\Spins\ring_{i}.m3d"
    doc.SaveAs(path)

print("Генерация завершена.")
```
